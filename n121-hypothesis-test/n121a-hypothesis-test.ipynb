## 1. One-Sample T-test Ïó∞Ïäµ

### Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
- [ÏÑúÏö∏Ïãú Í∞ÄÎ°úÏàò Îç∞Ïù¥ÌÑ∞ÏÖã](https://data.seoul.go.kr/dataList/367/S/2/datasetView.do)ÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.

Îã§Ïùå ÎßÅÌÅ¨Î•º Ï∞∏Ï°∞ÌïòÏÑ∏Ïöî. 
<https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt>


```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§
import pandas as pd

trees = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt', sep = '\t', skiprows = 1)

trees = trees.replace({'-':0})

trees
```

### "ÏÑúÏö∏ÏãúÏóêÎäî Íµ¨Î≥ÑÎ°ú ÌèâÍ∑† ÏïΩ 400 Í∑∏Î£® Ï†ïÎèÑÏùò `Ïù¥ÌåùÎÇòÎ¨¥`Í∞Ä ÏûàÎã§"ÎùºÎäî Í∞ÄÏÑ§Ïóê ÎåÄÌï¥ Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ ÏãúÌñâÌïòÏÑ∏Ïöî.

- Îã®, `ÏÑúÏö∏Ïãú`, `Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå`, `ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®` Î∂ÄÎ∂ÑÏùò Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§.
- Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalue Í∞íÏùÑ `pv1`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî. (`.pvalue` ÏÇ¨Ïö©)

```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§
import numpy as np

retusus = pd.DataFrame([trees.ÏûêÏπòÍµ¨, trees.Ïù¥ÌåùÎÇòÎ¨¥]).T.set_index('ÏûêÏπòÍµ¨').drop(['ÏÑúÏö∏Ïãú','Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå','ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®']).reset_index()

#SeriesÏùò Ï≤úÎã®ÏúÑÏΩ§Îßà Ï†úÍ±∞ Ìï®Ïàò
def ToNumericSeries (series):
  return pd.to_numeric(series.str.replace(',',''))

retusus.Ïù¥ÌåùÎÇòÎ¨¥ = ToNumericSeries(retusus.Ïù¥ÌåùÎÇòÎ¨¥)
#retusus.info()
'''
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 25 entries, 0 to 24
Data columns (total 2 columns):
 #   Column  Non-Null Count  Dtype 
---  ------  --------------  ----- 
 0   ÏûêÏπòÍµ¨     25 non-null     object
 1   Ïù¥ÌåùÎÇòÎ¨¥    25 non-null     int64 
dtypes: int64(1), object(1)
memory usage: 528.0+ bytes
'''
from scipy import stats

pv1 = stats.ttest_1samp(retusus.Ïù¥ÌåùÎÇòÎ¨¥,400).pvalue
print(retusus.Ïù¥ÌåùÎÇòÎ¨¥.mean())
print(pv1,"\n")

#ÏñëÏ∏°Í≤ÄÏ†ïÏóêÏÑú p-valueÏôÄ Ïú†ÏùòÏàòÏ§ÄÏùÑ ÌÜµÌï¥ ÌèâÍ∞ÄÎ•º returnÌïòÎäî Ìï®Ïàò
def reHypo_2tail(pval, SigLevel):
  if (SigLevel < pval) & (pval < 1-SigLevel):
    return "Ïú†ÏùòÏàòÏ§Ä : ["+str(SigLevel*100)+"%]\n\n"+"ÌèâÍ∞Ä     : p-value ="+str(round(pval,5))+"Ïù¥Í≥† Íµ¨Í∞Ñ("+str(SigLevel)+" < p-value < "+str(round(1-SigLevel,5))+")ÏùÑ ÎßåÏ°±ÌïòÏó¨\n           ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÎØÄÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïòÏßÄ ÏïäÎäîÎã§"
  else :
    return "Ïú†ÏùòÏàòÏ§Ä : ["+str(SigLevel*100)+"%]\n\n"+"ÌèâÍ∞Ä     : p-value ="+str(round(pval,5))+"Ïù¥Í≥† Íµ¨Í∞Ñ("+str(SigLevel)+" < p-value < "+str(round(1-SigLevel,5))+")ÏùÑ Î≤óÏñ¥ÎÇò\n           ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÏßÄ ÏïäÏúºÎØÄÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïúÎã§"

print("Í∑ÄÎ¨¥Í∞ÄÏÑ§ : ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† Ïù¥ÌåùÎÇòÎ¨¥Ïùò ÏàòÎäî ÏïΩ 400Í∑∏Î£®Ïù¥Îã§")
print("ÎåÄÏïàÍ∞ÄÏÑ§ : ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† Ïù¥ÌåùÎÇòÎ¨¥Ïùò ÏàòÎäî ÏïΩ 400Í∑∏Î£®Í∞Ä ÏïÑÎãàÎã§")
print(reHypo_2tail(pv1,0.05))
```

### Ïù¥Î≤àÏóêÎäî 10Í∞úÏùò Íµ¨Î•º ÏÉòÌîåÎßÅÌïòÏó¨, ÏúÑÏóê ÎåÄÌïú Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ Îã§Ïãú ÌïúÎ≤à ÏãúÌñâÌï¥Î≥¥ÏÑ∏Ïöî.

- Îã®, ÏïÑÎûòÏùò ÏΩîÎìúÎ•º Ï∞∏Ï°∞ÌïòÏó¨ sampleÏùÑ ÏúÑÌïú seed Îäî 123Î°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî.
- Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalueÍ∞íÏùÑ `pv2`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî


```python
np.random.seed(123)

SampleIndexNum = np.random.choice(len(retusus.Ïù¥ÌåùÎÇòÎ¨¥), size = 10)#,replace  = False) : Ï§ëÎ≥µÎ∂àÍ∞Ä, Í∏∞Î≥∏ÏùÄ Ï§ëÎ≥µÍ∞ÄÎä•
SampleRetusus = retusus.iloc[SampleIndexNum,1]

pv2 = stats.ttest_1samp(SampleRetusus,400).pvalue
print(retusus.Ïù¥ÌåùÎÇòÎ¨¥.mean())
print(pv2)

print("Í∑ÄÎ¨¥Í∞ÄÏÑ§ : ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† Ïù¥ÌåùÎÇòÎ¨¥Ïùò ÏàòÎäî ÏïΩ 400Í∑∏Î£®Ïù¥Îã§")
print("ÎåÄÏïàÍ∞ÄÏÑ§ : ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† Ïù¥ÌåùÎÇòÎ¨¥Ïùò ÏàòÎäî ÏïΩ 400Í∑∏Î£®Í∞Ä ÏïÑÎãàÎã§")
print(reHypo_2tail(pv2,0.05))
```

## 2. Two-Sample T-test Ïó∞Ïäµ

### "ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥ÏàòÏôÄ ÏôïÎ≤öÎÇòÎ¨¥Ïùò ÏàòÎäî Ï∞®Ïù¥Í∞Ä ÏóÜÎã§"ÎùºÎäî Í∞ÄÏÑ§Ïóê ÎåÄÌï¥ Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ ÏãúÌñâÌïòÏÑ∏Ïöî.

- Îã®, `ÏÑúÏö∏Ïãú`, `Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå`, `ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®` Î∂ÄÎ∂ÑÏùò Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§.
- Í∞ÄÏÑ§Í≤ÄÏ†ï Í≤∞Í≥ºÏùò pvalueÎ•º `pv3`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî

```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§
TreeComp = pd.DataFrame([trees.ÏûêÏπòÍµ¨, trees.ÎäêÌã∞ÎÇòÎ¨¥, trees.ÏôïÎ≤öÎÇòÎ¨¥]).T.set_index('ÏûêÏπòÍµ¨').drop(['ÏÑúÏö∏Ïãú','Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå','ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®']).reset_index()
TreeComp.ÎäêÌã∞ÎÇòÎ¨¥ = ToNumericSeries(TreeComp.ÎäêÌã∞ÎÇòÎ¨¥)
TreeComp.ÏôïÎ≤öÎÇòÎ¨¥ = ToNumericSeries(TreeComp.ÏôïÎ≤öÎÇòÎ¨¥)
#TreeComp.info()
'''
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 25 entries, 0 to 24
Data columns (total 3 columns):
 #   Column  Non-Null Count  Dtype 
---  ------  --------------  ----- 
 0   ÏûêÏπòÍµ¨     25 non-null     object
 1   ÎäêÌã∞ÎÇòÎ¨¥    25 non-null     int64 
 2   ÏôïÎ≤öÎÇòÎ¨¥    25 non-null     int64 
dtypes: int64(2), object(1)
memory usage: 728.0+ bytes
'''
pv3 = stats.ttest_ind(TreeComp.ÎäêÌã∞ÎÇòÎ¨¥,TreeComp.ÏôïÎ≤öÎÇòÎ¨¥).pvalue
print(TreeComp.ÎäêÌã∞ÎÇòÎ¨¥.mean(), TreeComp.ÏôïÎ≤öÎÇòÎ¨¥.mean(),"\n",pv3,"\n")

print(r"Í∑ÄÎ¨¥Í∞ÄÏÑ§ : 'ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥ Ïàò'ÏôÄ 'ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ Ïàò' Îäî Ï∞®Ïù¥Í∞Ä ÏóÜÎã§")
print(r"ÎåÄÏïàÍ∞ÄÏÑ§ : 'ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥ Ïàò'ÏôÄ 'ÏÑúÏö∏Ïãú Íµ¨Î≥Ñ ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ Ïàò' Îäî Ï∞®Ïù¥Í∞Ä ÏûàÎã§")
print(reHypo_2tail(pv3,0.05))
```


### Two-Sample One-tailed testÎ•º ÏÇ¨Ïö©ÌïòÏó¨ **ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥ÏàòÍ∞Ä ÏôïÎ≤öÎÇòÎ¨¥ ÏàòÎ≥¥Îã§ ÌÅ¨Îã§**ÎùºÎäî Í∞ÄÏÑ§Ïóê ÎåÄÌïú Í∞ÄÏÑ§Í≤ÄÏ†ïÏùÑ ÏãúÌñâÌïòÏÑ∏Ïöî.
- Ìï¥Îãπ Í∞ÄÏÑ§Í≤ÄÏ†ïÏùò Í≤∞Í≥ºÎ•º `ans`Ïóê Ï†ÄÏû•ÌïòÏÑ∏Ïöî.

Ïù¥ Îïå `ans`Í∞Ä Í∞ÄÏßà Ïàò ÏûàÎäî Í∞íÏùÄ ÏïÑÎûòÏôÄ Í∞ôÏäµÎãàÎã§ :
1. "ÎäêÌã∞ÎÇòÎ¨¥" ( ÎäêÌã∞ÎÇòÎ¨¥Ïùò ÏàòÍ∞Ä ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÍ≤å ÎßéÏùÄ Í≤ΩÏö∞ )
2. "ÏôïÎ≤öÎÇòÎ¨¥" ( ÏôïÎ≤öÎÇòÎ¨¥Ïùò ÏàòÍ∞Ä ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÍ≤å ÎßéÏùÄ Í≤ΩÏö∞ )
3. "ÏóÜÏùå" (ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÏßÄ ÏïäÏùÄ Í≤ΩÏö∞)

Two-Sample One-tailed testÎ•º scipyÏóêÏÑú ÌïòÎäî Î∞©Î≤ïÏùÄ ÏïÑÎûòÏùò ÎßÅÌÅ¨Î•º Ï∞∏Í≥† ÌïòÏÑ∏Ïöî  
Ï∞∏Í≥† ÎßÅÌÅ¨ : https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

```python
# ÏïÑÎûòÏùò ÏΩîÎìúÎ•º Ïã§ÌñâÌï¥Ï£ºÏÑ∏Ïöî
!pip install scipy --upgrade
```

```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§

#Two-Sample One-tailed test : Ïù¥Ìïò 'Îã®Ï∏°Í≤ÄÏ†ï' ÏúºÎ°ú ÌëúÍ∏∞Ìï®

#pvalue(list, len = 2 : Îã®Ï∏°Í≤ÄÏ†ïÏùò Îëê Ìï≠Î™©Ïóê ÎåÄÌïú P-valueÍ∞í)Í∞Ä ÏùòÎØ∏ÌïòÎäî Ìï≠Î™©ÏùÑ return ÌïòÎäî Ìï®Ïàò
def reAnswerToTree (pval,SigLevel) :
  #ÎäêÌã∞ÎÇòÎ¨¥ != ÏôïÎ≤öÎÇòÎ¨¥(Í∑ÄÎ¨¥Í∞ÄÏÑ§Ïù¥ Ï∞∏Ïù¥ ÏïÑÎãò) Ïù¥Î©¥ÏÑú ÎäêÌã∞ÎÇòÎ¨¥ > ÏôïÎ≤öÎÇòÎ¨¥ÏùºÎïå (alternative = 'greater' : ÎåÄÎ¶ΩÍ∞ÄÏÑ§Ïù¥ upper tail)
  if pval[0] < SigLevel:
    return "ÎäêÌã∞ÎÇòÎ¨¥"
  #ÎäêÌã∞ÎÇòÎ¨¥ != ÏôïÎ≤öÎÇòÎ¨¥(Í∑ÄÎ¨¥Í∞ÄÏÑ§Ïù¥ Ï∞∏Ïù¥ ÏïÑÎãò) Ïù¥Î©¥ÏÑú ÎäêÌã∞ÎÇòÎ¨¥ < ÏôïÎ≤öÎÇòÎ¨¥ÏùºÎïå (alternative = 'less' : ÎåÄÎ¶ΩÍ∞ÄÏÑ§Ïù¥ lower tail)
  elif pval[1] < SigLevel :
    return "ÏôïÎ≤öÎÇòÎ¨¥"
  else :
    return "ÏóÜÏùå"

#Îã®Ï∏°Í≤ÄÏ†ï Ìï®Ïàò
def reHypo_1tail(pval, SigLevel):
  if SigLevel < pval:
    return "Ïú†ÏùòÏàòÏ§Ä : ["+str(SigLevel*100)+"%]\n\n"+"ÌèâÍ∞Ä     : p-value ="+str(round(pval,5))+"Ïù¥Í≥† Íµ¨Í∞Ñ("+str(SigLevel)+" < p-value < 1)ÏùÑ ÎßåÏ°±ÌïòÏó¨\n           ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÎØÄÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïòÏßÄ ÏïäÎäîÎã§"
  else :
    return "Ïú†ÏùòÏàòÏ§Ä : ["+str(SigLevel*100)+"%]\n\n"+"ÌèâÍ∞Ä     : p-value ="+str(round(pval,5))+"Ïù¥Í≥† Íµ¨Í∞Ñ("+str(SigLevel)+" < p-value < 1)ÏùÑ Î≤óÏñ¥ÎÇò\n           ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÌïòÏßÄ ÏïäÏúºÎØÄÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïúÎã§"

#p_value Ïóê Îã®Ï∏°Í≤ÄÏ†ï(upper tail/lower tail)Ïùò p-valueÍ∞íÏùÑ Ï†ÄÏû•
p_value = [stats.ttest_ind(TreeComp.ÎäêÌã∞ÎÇòÎ¨¥,TreeComp.ÏôïÎ≤öÎÇòÎ¨¥,alternative = 'greater').pvalue,
           stats.ttest_ind(TreeComp.ÎäêÌã∞ÎÇòÎ¨¥,TreeComp.ÏôïÎ≤öÎÇòÎ¨¥,alternative = 'less').pvalue]
print("Ìï≠Î™©    : ÎäêÌã∞ÎÇòÎ¨¥            ÏôïÎ≤öÎÇòÎ¨¥",
      "\nÌèâÍ∑†    :",TreeComp.ÎäêÌã∞ÎÇòÎ¨¥.mean(),"           ",TreeComp.ÏôïÎ≤öÎÇòÎ¨¥.mean(),
      "\nP-value :",p_value[0],p_value[1],"\n")
ans = reAnswerToTree(p_value,0.05)


print("\n1.Îã®Ï∏°Í≤ÄÏ†ï(upper tail : Ïö∞Ï∏°)")
print("Í∑ÄÎ¨¥Í∞ÄÏÑ§ : 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥Ïàò'Í∞Ä 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ Ïàò'ÏôÄ Í∞ôÎã§")
print("ÎåÄÎ¶ΩÍ∞ÄÏÑ§ : 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥Ïàò'Í∞Ä 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ Ïàò'Î≥¥Îã§ ÌÅ¨Îã§")
print(reHypo_1tail(p_value[0],0.05))
print("\n2.Îã®Ï∏°Í≤ÄÏ†ï(lower tail : Ï¢åÏ∏°)")
print("Í∑ÄÎ¨¥Í∞ÄÏÑ§ : 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥Ïàò'Í∞Ä 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ Ïàò'ÏôÄ Í∞ôÎã§")
print("ÎåÄÎ¶ΩÍ∞ÄÏÑ§ : 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÎäêÌã∞ÎÇòÎ¨¥Ïàò'Í∞Ä 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÏôïÎ≤öÎÇòÎ¨¥ Ïàò'Î≥¥Îã§ ÏûëÎã§")
print(reHypo_1tail(p_value[1],0.05))
print("\n----------------------------------------------------------------\n")
print("ÏµúÏ¢ÖÍ≤∞Î°† : 'ÏÑúÏö∏ÏãúÏùò Íµ¨Î≥Ñ ÌèâÍ∑† ÎÇòÎ¨¥Ïùò Ïà´Ïûê ÎπÑÍµê'ÏóêÏÑú\n           ÎäêÌã∞ÎÇòÎ¨¥ Ìï≠Î™©Í≥º ÏôïÎ≤öÎÇòÎ¨¥ Ìï≠Î™© Ï§ë Îçî ÌÅ¨Îã§Í≥† Ìï† Ïàò ÏûàÎäî Ìï≠Î™©ÏùÄ ->",ans)

```

## üî• **ÎèÑÏ†ÑÍ≥ºÏ†ú**

### 1. Compare mean for multiple-sample 
 
Îã§Ïùå ÎßÅÌÅ¨Î•º Ï∞∏Ï°∞ÌïòÏÑ∏Ïöî. <https://reneshbedre.github.io/blog/anova.html>

Íµ¨Î≥Ñ, ÏùÄÌñâÎÇòÎ¨¥, ÏñëÎ≤ÑÏ¶òÎÇòÎ¨¥, ÎäêÌã∞ÎÇòÎ¨¥Ïóê ÎåÄÌï¥ÏÑú ANOVA TestÎ•º Ïã§ÌñâÌïòÏÑ∏Ïöî.

Îã®, `ÏÑúÏö∏Ïãú`, `Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå`, `ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®` Î∂ÄÎ∂ÑÏùò Îç∞Ïù¥ÌÑ∞Îäî Ï†úÏô∏Ìï©ÎãàÎã§.

```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§

#ÏÉà DataFrame ÏÉùÏÑ±ÌïòÍ∏∞ : treesofAnovaTest / ÏàòÎüâ Ìï≠Î™©Îì§ Ï†ïÎ¶¨(Ï≤úÎã®ÏúÑÏΩ§Îßà Ï†úÍ±∞, numeric Î≥ÄÌôò)
treesofAnovaTest = pd.DataFrame([trees.ÏûêÏπòÍµ¨, trees.ÏùÄÌñâÎÇòÎ¨¥, trees.ÏñëÎ≤ÑÏ¶òÎÇòÎ¨¥, trees.ÎäêÌã∞ÎÇòÎ¨¥]).T.set_index('ÏûêÏπòÍµ¨').drop(['ÏÑúÏö∏Ïãú','Í≥µÏõêÎÖπÏßÄÏÇ¨ÏóÖÏÜå','ÏãúÏÑ§Í¥ÄÎ¶¨Í≥µÎã®']).reset_index()
treesofAnovaTest.ÏùÄÌñâÎÇòÎ¨¥ = ToNumericSeries(treesofAnovaTest.ÏùÄÌñâÎÇòÎ¨¥)
treesofAnovaTest.ÏñëÎ≤ÑÏ¶òÎÇòÎ¨¥ = ToNumericSeries(treesofAnovaTest.ÏñëÎ≤ÑÏ¶òÎÇòÎ¨¥)
treesofAnovaTest.ÎäêÌã∞ÎÇòÎ¨¥ = ToNumericSeries(treesofAnovaTest.ÎäêÌã∞ÎÇòÎ¨¥)
display(treesofAnovaTest.head())

#melt Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ ÏÉàÎ°≠Í≤å Îç∞Ïù¥ÌÑ∞Î•º Ï†ïÎ¶¨ÌïòÍ∏∞(ÏûêÏπòÍµ¨ Í∏∞Ï§ÄÏúºÎ°ú Í∞Å ÎÇòÎ¨¥Ïùò ÏàòÎüâ Ï†ïÎ¶¨)
treesofAnovaTest_melt = pd.melt(treesofAnovaTest, id_vars=['ÏûêÏπòÍµ¨'], value_vars=['ÏùÄÌñâÎÇòÎ¨¥', 'ÏñëÎ≤ÑÏ¶òÎÇòÎ¨¥', 'ÎäêÌã∞ÎÇòÎ¨¥'])
treesofAnovaTest_melt.columns = ['ÏûêÏπòÍµ¨', 'ÎÇòÎ¨¥', 'ÏàòÎüâ']
display(treesofAnovaTest_melt.head())
```

```python
#Anova Test Ï†ÅÏö©Ìï¥Î≥¥Í∏∞

# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(treesofAnovaTest.ÏùÄÌñâÎÇòÎ¨¥,treesofAnovaTest.ÏñëÎ≤ÑÏ¶òÎÇòÎ¨¥,treesofAnovaTest.ÎäêÌã∞ÎÇòÎ¨¥)
print(fvalue, pvalue)

# get ANOVA table as R like output
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ordinary Least Squares (OLS) model
model = ols('ÏàòÎüâ ~ C(ÎÇòÎ¨¥)', data=treesofAnovaTest_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
display(anova_table)
```

### 2. Sampling

- ÏÑ∏ÏÖò ÏãúÍ∞ÑÏóê Î∞∞Ïõ†Îçò 4Í∞ÄÏßÄ sampling (Simple Random, Systematic, Stratified Random, Cluster) Î∞©ÏãùÏùÑ Íµ¨ÌòÑÌï¥Î≥¥ÏÑ∏Ïöî.

- Ïù¥Î•º ÏúÑÌïú Îç∞Ïù¥ÌÑ∞Îäî 1 Î∂ÄÌÑ∞ 100ÏùÑ Îã¥ÏùÄ numpy arrayÏù¥Î©∞, Sampling Ìï¥Ïïº ÌïòÎäî ÏàòÎäî 20Í∞ú ÏûÖÎãàÎã§.

- Stratified Random SamplingÏùÑ ÏúÑÌïú Í∑∏Î£πÏùÑ ÎÇòÎàÑÎäî Í∏∞Ï§ÄÏùÄ 1-10, 11-20, ... Í≥º Í∞ôÏùÄ Î∞©ÏãùÏûÖÎãàÎã§.

- Cluster SamplingÏóêÏÑú Í∑∏Î£πÏùÑ ÎÇòÎàÑÎäî Í∏∞Ï§ÄÏùÄ 5Î°ú ÎÇòÎà¥ÏùÑÎïåÏùò ÎÇòÎ®∏ÏßÄ Í∞íÏûÖÎãàÎã§. (0,1,2,3,4)

```python
# Ïù¥ Í≥≥Ïóê ÎãµÏïàÏùÑ ÏûëÏÑ±ÌïòÏãúÍ∏∞ Î∞îÎûçÎãàÎã§

#Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± / ÏÉòÌîå ÌÅ¨Í∏∞ ÏßÄÏ†ï
data = np.arange(1,101)
SamplingSize = 20


#1. Simple Random

SimpleRandomSampling = list()
for i in range(SamplingSize) :
  SimpleRandomSampling = SimpleRandomSampling + data[np.random.choice(len(data), size = 1)].tolist() #Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏù¥ np.array


#2. Systematic

SystematicSampling = list()
#[1, 6, 11, 16,...,96]
for i in range(SamplingSize) :
  SystematicSampling.append(data[int(len(data)/SamplingSize)*i]) #Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏù¥ intÏù∏ Îç∞Ïù¥ÌÑ∞Ïù¥ÎØÄÎ°ú appendÏÇ¨Ïö©


#3. Stratified Random

StratifiedRandomSampling = list()
#Ï∂îÏ∂ú Íµ¨Í∞ÑÎ≥Ñ Ìè¨Ìï® Îç∞Ïù¥ÌÑ∞ Í∞ØÏàò
SamplingGroupData = 2
for i in range(int(SamplingSize/SamplingGroupData)) :
  for _ in range(SamplingGroupData):
    StratifiedRandomSampling = StratifiedRandomSampling + data[np.random.choice(int(SamplingSize/SamplingGroupData)*(i+1),size=1)].tolist()


#4. Cluster

ClusterSampling = list()
#SamplingNÏúºÎ°ú ÎÇòÎà¥ÏùÑ Îïå ÎÇòÎ®∏ÏßÄ(ReminderSamplingGroup)Î•º Í∑∏Î£πÏúºÎ°ú ÏÑ†Ï†ïÌïòÍ∏∞(0, 1, 2, 3, 4)
SamplingN = 5
ReminderSamplingGroup = 2

for i in range(SamplingSize):
  ClusterSampling = ClusterSampling + data[np.random.choice(int(len(data)/SamplingN),size = 1)*SamplingN+ReminderSamplingGroup].tolist()


#Ï∂úÎ†•ÌïòÍ∏∞
print("1. Simple Random\n",SimpleRandomSampling,
      "\n2. Systematic\n",SystematicSampling,
      "\n3. Stratified Random\n",StratifiedRandomSampling,
      "\n4. Cluster\n",ClusterSampling)
```
